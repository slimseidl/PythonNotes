```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create array for x- and y- coordinates
x = [1, 2, 3, 4]
y = [5, 1, 4, 2]
# Create a 6x5 figure only need to use plt.figure when changing dimension size from default (most of the time)
plt.figure(figsize=[6, 5])
# Create a line plot
plt.plot(x, y)
```

```python
# Create an equally spaced array with 100
# elements from -5 to 5
x = np.linspace(-5, 5, num=100)
# Plots the graph for y = x^2
plt.plot(x, x * x)
# Plots the graph for y = -4x - 4
plt.plot(x, -4 * x - 4)
```
