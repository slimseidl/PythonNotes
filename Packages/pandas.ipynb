```python
# Country Data
country = pd.read_csv('country.csv')
```

```python
country
```

```python
# Select rows where the Continent is South America
country[country['Continent'] == 'South America']
```

```python
# Select rows where the Continent is not Asia
country[country['Continent'] != 'Asia']
```

```python
# Select rows where the Continent is Asia or Europe
country[(country['Continent'] == 'Asia') | (country['Continent'] == 'Europe')]
```

```python
# Select rows where the Continent is not in Asia
country[~(country['Continent'] != 'Asia')]
```

```python
# Using loc to display all rows after 2 and the tip column
cabsNY = pd.read_csv('taxi.csv')
df = cabsNY.loc[2:,['tip']]
df
```

```python
# Creating copy of df where tolls <= 10
taxicabsNY = pd.read_csv('taxi.csv')
df = taxicabsNY[taxicabsNY['tolls'] <= 10]
df.shape
```

```python
# Using multiple logical operators with .loc method 
taxizones = pd.read_csv('taxizones.csv')
drop_pick = taxizones.loc[(taxizones['dropoff_borough'] == 'Brooklyn') & ((taxizones['pickup_zone'] == 'LaGuardia Airport') | (taxizones['pickup_zone'] == 'Briarwood/Jamaica Hills'))]
```
